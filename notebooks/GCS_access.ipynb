```python
from google.cloud import storage
import google
import pandas as pd
import os
import io
```

```python
bucket_name = os.environ.get("BUCKET_NAME")
print(bucket_name)
client = storage.Client()
bucket = client.get_bucket(bucket_name)
order2016 = bucket.blob("orders2016.csv")
print(order2016)
```

```python
csv_data = order2016.download_as_string()
df = pd.read_csv(io.BytesIO(csv_data), sep=";")
```
